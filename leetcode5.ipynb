```python
# Solution for LeetCode 5: Longest Palindromic Substring
def longest_palindrome(s: str) -> str:
    if not s:
        return ""
    start, end = 0, 0

    def expand(l: int, r: int) -> tuple:
        while l >= 0 and r < len(s) and s[l] == s[r]:
            l -= 1
            r += 1
        return l + 1, r - 1  # inclusive indices of palindrome

    for i in range(len(s)):
        l1, r1 = expand(i, i)      # odd-length center
        l2, r2 = expand(i, i + 1)  # even-length center

        if (r1 - l1) > (end - start):
            start, end = l1, r1
        if (r2 - l2) > (end - start):
            start, end = l2, r2

    return s[start:end + 1]


# Quick manual tests
if __name__ == "__main__":
    samples = ["babad", "cbbd", "a", "", "forgeeksskeegfor"]
    for t in samples:
        print(t, "->", longest_palindrome(t))
```
