```python
def myPow(x: float, n: int) -> float:
    
    def power(base, exp):
        if exp == 0:
            return 1.0
        
        if exp % 2 == 0:
            half = power(base, exp // 2)
            return half * half
        else:
            return base * power(base, exp - 1)
            
    if n < 0:
        x = 1 / x
        n = -n
        
    return power(x, n)
```
