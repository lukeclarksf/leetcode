```python
# LeetCode 13 — Roman to Integer

# Problem: Convert a Roman numeral string to an integer.

# Approach: Map each Roman char to its value. Iterate left-to-right; if a value is less than the next value, subtract it, otherwise add it.

# Complexity: O(n) time, O(1) extra space.

# Example: "MCMXCIV" -> 1994

class Solution:
    def romanToInt(self, s: str) -> int:
        vals = {'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000}
        total = 0
        n = len(s)
        for i, ch in enumerate(s):
            v = vals[ch]
            if i + 1 < n and vals[s[i + 1]] > v:
                total -= v
            else:
                total += v
        return total

# Example usage:
sol = Solution()
print(sol.romanToInt("MCMXCIV"))  # Output: 1994

# Test cases
assert sol.romanToInt("III") == 3
assert sol.romanToInt("IV") == 4
assert sol.romanToInt("IX") == 9
assert sol.romanToInt("LVIII") == 58
assert sol.romanToInt("MCMXCIV") == 1994
print("All test cases passed!")
```
