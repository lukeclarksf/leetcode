```python
import math

def getPermutation(n: int, k: int) -> str:
    
    numbers = list(range(1, n + 1))
    output = []
    k -= 1
    
    for i in range(n, 0, -1):
        
        fact = math.factorial(i - 1)
        
        index = k // fact
        
        output.append(str(numbers[index]))
        
        numbers.pop(index)
        
        k %= fact
        
    return "".join(output)
```
