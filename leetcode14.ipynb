```python
from typing import List

# LeetCode 14 - Longest Common Prefix

class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        if not strs:
            return ""
        prefix_chars = []
        for chars in zip(*strs):  # iterate column-wise
            # if all characters in this column are the same, keep it
            if all(c == chars[0] for c in chars):
                prefix_chars.append(chars[0])
            else:
                break
        return "".join(prefix_chars)

# Quick smoke tests
if __name__ == "__main__":
    sol = Solution()
    print(sol.longestCommonPrefix(["flower","flow","flight"]))  # "fl"
    print(sol.longestCommonPrefix(["dog","racecar","car"]))     # ""
    print(sol.longestCommonPrefix([]))                         # ""
    print(sol.longestCommonPrefix(["a"]))                      # "a"
```
