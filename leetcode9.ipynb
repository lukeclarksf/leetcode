```python
# LeetCode 9: Palindrome Number
# New Jupyter cell (index 0) content

def is_palindrome(x: int) -> bool:
    """
    Return True if x is a palindrome integer without converting to string.
    Uses the 'reverse half' technique to avoid overflow and extra work.
    """
    if x < 0 or (x % 10 == 0 and x != 0):
        return False

    rev_half = 0
    while x > rev_half:
        rev_half = rev_half * 10 + x % 10
        x //= 10

    # For even number of digits: x == rev_half
    # For odd number of digits: x == rev_half // 10 (middle digit ignored)
    return x == rev_half or x == rev_half // 10


# LeetCode-style class (optional)
class Solution:
    def isPalindrome(self, x: int) -> bool:
        return is_palindrome(x)


# Quick tests
if __name__ == "__main__":
    tests = [121, -121, 10, 0, 12321, 123321, 1]
    for t in tests:
        print(t, is_palindrome(t))
```
