```python
def countAndSay(n: int) -> str:
    if n == 1:
        return "1"
    
    prev = "1"
    
    for _ in range(2, n + 1):
        current = ""
        i = 0
        while i < len(prev):
            count = 1
            char = prev[i]
            
            while i + 1 < len(prev) and prev[i] == prev[i+1]:
                count += 1
                i += 1
                
            current += str(count) + char
            i += 1
            
        prev = current
        
    return prev
```
