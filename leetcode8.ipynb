```python
# leetcode8_myAtoi.py
class Solution:
    def myAtoi(self, s: str) -> int:
        s = s.lstrip()
        if not s:
            return 0
        sign = 1
        i = 0
        if s[0] in '+-':
            if s[0] == '-':
                sign = -1
            i = 1
        res = 0
        INT_MAX = 2**31 - 1
        INT_MIN = -2**31
        while i < len(s) and s[i].isdigit():
            res = res * 10 + (ord(s[i]) - 48)
            if sign * res > INT_MAX:
                return INT_MAX
            if sign * res < INT_MIN:
                return INT_MIN
            i += 1
        return sign * res

if __name__ == "__main__":
    sol = Solution()
    tests = [
        ("42", 42),
        ("   -42", -42),
        ("4193 with words", 4193),
        ("words and 987", 0),
        ("-91283472332", -2**31),
        ("91283472332", 2**31 - 1),
        ("+1", 1),
        ("", 0),
        ("   +0 123", 0),
    ]
    for inp, expected in tests:
        assert sol.myAtoi(inp) == expected, f"{inp!r} -> {sol.myAtoi(inp)} != {expected}"
    print("All tests passed.")
```
