```python
def reverse(x: int) -> int:
    INT_MAX = 2**31 - 1
    INT_MIN = -2**31

    sign = 1 if x >= 0 else -1
    x = abs(x)
    res = 0

    while x:
        pop = x % 10
        x //= 10
        # check overflow before pushing digit
        if res > INT_MAX // 10 or (res == INT_MAX // 10 and pop > INT_MAX % 10):
            return 0
        res = res * 10 + pop

    return sign * res


# Example tests
print(reverse(123))         # 321
print(reverse(-123))        # -321
print(reverse(120))         # 21
print(reverse(1534236469))  # 0 (overflow)
```
