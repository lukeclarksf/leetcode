```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        seen = {}
        for i, num in enumerate(nums):
            need = target - num
            if need in seen:
                return [seen[need], i]
            seen[num] = i
        raise ValueError("No two sum solution")

# Example usage
nums = [2, 7, 11, 15]
target = 9
print(Solution().twoSum(nums, target))  # [0, 1]
```
