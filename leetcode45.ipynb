```python
def jump(nums: list[int]) -> int:
    n = len(nums)
    if n <= 1:
        return 0
        
    jumps = 0
    current_reach = 0
    farthest_reach = 0
    
    for i in range(n - 1):
        farthest_reach = max(farthest_reach, i + nums[i])
        
        if i == current_reach:
            jumps += 1
            current_reach = farthest_reach
            
            if current_reach >= n - 1:
                break
                
    return jumps
```
