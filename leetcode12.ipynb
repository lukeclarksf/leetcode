```python
# LeetCode 12: Integer to Roman
# Place this cell at index 0 of the notebook.

class Solution:
    def intToRoman(self, num: int) -> str:
        vals = [
            (1000, "M"), (900, "CM"), (500, "D"), (400, "CD"),
            (100, "C"), (90, "XC"), (50, "L"), (40, "XL"),
            (10, "X"), (9, "IX"), (5, "V"), (4, "IV"), (1, "I")
        ]
        res = []
        for v, s in vals:
            if num == 0:
                break
            count, num = divmod(num, v)
            res.append(s * count)
        return "".join(res)

# Quick checks
sol = Solution()
assert sol.intToRoman(3) == "III"
assert sol.intToRoman(4) == "IV"
assert sol.intToRoman(9) == "IX"
assert sol.intToRoman(58) == "LVIII"  # L=50, V=5, III=3
assert sol.intToRoman(1994) == "MCMXCIV"
print("All tests passed.")
```
